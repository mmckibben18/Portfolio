```python
# Load dataset
adult_income = pd.read_csv('adult_income_data.csv')
```

```python
# Look at dataframe structure
adult_income.head(2)
```

```python
# Create empty list
header = []
# Use with to open and close the txt file 
with open('adult_income_names.txt', 'r') as headers:
    # Read lines of text document
    for line in headers:
        headers.readline()
        # Separate the first line based on ":" 
        topline = line.split(":")[0]
        # Add the split values to the list
        header.append(topline)
        pass
header
```

```python
# Add income to the list
header.append('Income')
```

```python
# Reload the data this time with the header list as the headers
adult_income = pd.read_csv('adult_income_data.csv', names = header)
# Look at the df
adult_income.head()
```

```python
# Find all missing values and add them for each series
missing_vals = adult_income.isnull().sum()
```

```python
# Display missing value count
missing_vals
```

```python
# Subset df
small_income = adult_income[['age', 'education', 'occupation']]
```

```python
# Look at structure of subset df
small_income.head()
```

```python
# Plot histogram of age column
plt.hist(small_income['age'], bins = 20)
plt.xlim(0, 100)
plt.show
```

```python
# Define a function to remove whitespace from string
# Takes one parameter
def remove_whitespace(string):
    string.replace(" ", "")
    return string
```

```python
# Gather some info on df series
adult_income.dtypes
```

```python
# Use apply to remove whitespace with our function
adult_income.apply(remove_whitespace)
```

```python
# Find the entries that aged 30-50
thirty_to_fifty = adult_income[(adult_income['age']>29) & (adult_income['age']<51)].count()
```

```python
# See the number of values that meet our criteria
thirty_to_fifty
```

```python
# Group by age and education
age_dist_education = adult_income.groupby(['age', 'education'])
# Get summary stats on df in new df
age_desc_education = age_dist_education.describe()
```

```python
# Show df
age_desc_education.head(18)
```

```python
# Plot age distribution by education level and make the figure bigger
adult_income.boxplot(column = 'age', by = 'education', figsize = (28,18))
# Make the font bigger and rotate to fit better
plt.xticks(fontsize = 38, rotation = 90)
# Make font bigger
plt.yticks(fontsize = 38)
# Give it a large title
plt.title(label = 'Age Distribution for Education', fontsize = 48)
# Show the plot
plt.show()
```

```python
# Get age distribution for occupation
age_dist_occupation = adult_income.groupby('occupation')
# Get summary statistics for occupation
age_desc_occupation = age_dist_occupation.describe()
```

```python
# Display the summary statistics for age
age_desc_occupation['age']
```

```python
# Find the oldest average occupation value
age_desc_occupation['age']['mean'].max()
```
