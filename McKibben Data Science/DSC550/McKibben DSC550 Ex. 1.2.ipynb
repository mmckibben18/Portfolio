```python
# Finds all the publishers
publishers = df["Publisher"].unique()
```

```python
# Find the five publishers with the highest total global sales
z = []
for x in publishers:
    pub_sales = df[df.Publisher == x]["Global_Sales"].sum()
    print(x, " : ", pub_sales)
    z.append(x)
    z.append(pub_sales)
z    
```

```python
y = {z[i]: z[i+1] for i in range(0, len(z), 2)}
{k: v for k, v in sorted(y.items(), key = lambda item: item[1])}
# Last five values are the top five publishers of global sales
```
