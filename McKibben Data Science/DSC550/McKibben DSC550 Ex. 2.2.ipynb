##### The dataset we have chosen provides some interesting insights into the relative prevalence of schizophrenia in various countries. We started with the US, the UK, Canada, Australia, and New Zealand. All are first-world countries but regionally, culturally, and politically different. The US, New Zealand, Australia, and Canada have the highest prevalence rates out of all ten countries, from 0.310 - 0.375%. The UK comes in at the lowest in this group, with its rate peaking at 0.260%. Another interesting observation is that the UK is the only place to show a steady rise in prevalence from 1990 to 2016. In 2016, the US was decreasing, and New Zealand remained relatively stable. We do not have much information about the origins of the dataset, so we're unsure of what we could attribute the differences in the trends to. We'd like to posit the theory though that because the US, Canada, Australia, and New Zealand all use the DSM-5 to diagnose mental illnesses that there is a higher rate of occurence compared to the countries that use the ICD-10 due to the diagnostic criteria.

##### We also looked at some Nordic countries: Finland, Sweden, Denmark, and Iceland. The plot for Finland, which was once a territory of Sweden, looks very similar to the Swedish plot. The plots start at their highest points and decline, then decline steeply starting in 2010. Sweden has the highest peak percentage of the Nordic countries at 0.282%, and Finland's highest point is 0.270%. Meanwhile, Denmark starts at its lowest rate, then spikes from 2005 to 2010, and Iceland is unlike the previous plots in that it has two local minima. The peak prevalence percentage for Iceland and Denmark is roughly 0.256%, the lowest of all ten plots.

##### Japan is middling with a rate of 0.304%. It was chosen to provide another reference point that was not Nordic, nor using the DSM-5.

##### Note that changing the scale of the y-axis changes the look and story the plots tell. When plotting all the data on the same chart with a y-axis that accommodates the breadth of the data, some of the dramatic features present on a much smaller range of the y-axis are removed. For instance, the dual local minima from Iceland's plot are nearly imperceptible. New Zealand again appears relatively stable over time. Japan and Denmark finished at a higher rate than when they started. Finland and Sweden still show a downward slope, but it is much less pronounced on this larger scale. The upward trend from the UK is much less drastic in the final plot than the plot of the UK on its own. Lastly, the cosinusoidal appearance of the rate from the US is still present but muted.
