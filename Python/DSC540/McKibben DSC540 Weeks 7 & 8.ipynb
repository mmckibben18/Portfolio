```python
# Define the function to remove punctuation
def remove_punc(text):
    text = text.translate(str.maketrans('', '', string.punctuation))
    return text
```

```python
# Use the apply function to remove punctuation
df['stripped repository'] = df['Repository'].apply(lambda x: remove_punc(x) if type(x) == str else x)
df
```
